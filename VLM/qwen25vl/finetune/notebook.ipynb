```python
import ast

with open('/purestorage/AILAB/AI_1/dataset/Objects365/labels/train/objects365_v1_00000000.txt', "r") as f:
    data = f.read()
    lines = ast.literal_eval(data)  #     list 
```

```python
for line in lines:
    parts = line.split(' ')
    if len(parts) != 5:
        continue

    class_id, x_center, y_center, box_width, box_height = map(float, parts)

    #     
    x_center *= width
    y_center *= height
    box_width *= width
    box_height *= height


    #    
    x0 = max(x_center - box_width / 2, 0)
    y0 = max(y_center - box_height / 2, 0)
    x1 = min(x_center + box_width / 2, width)
    y1 = min(y_center + box_height / 2, height)

    print(x0, y0, x1, y1)
    #  
    draw.rectangle([x0, y0, x1, y1], outline="red", width=10)

    draw.text((x0 + 2, y0 + 2), str(int(class_id)), fill="yellow")
```

```python
from PIL import Image, ImageDraw, ImageFont
import os

# 
image_path = "/purestorage/AILAB/AI_1/dataset/Objects365/images/train/objects365_v1_00000000.jpg"
label_path = "/purestorage/AILAB/AI_1/dataset/Objects365/labels/train/objects365_v1_00000000.txt"

#  
image = Image.open(image_path).convert("RGB")
width, height = image.size
draw = ImageDraw.Draw(image)

#   
with open(label_path, "r") as f:
    lines = f.readlines()

for line in lines:
    parts = line.strip().split()
    if len(parts) != 5:
        continue

    class_id, x_center, y_center, box_width, box_height = map(float, parts)

    #      
    x_center *= width
    y_center *= height
    box_width *= width
    box_height *= height

    x0 = max(x_center - box_width / 2, 0)
    y0 = max(y_center - box_height / 2, 0)
    x1 = min(x_center + box_width / 2, width)
    y1 = min(y_center + box_height / 2, height)

    #   ( )
    draw.rectangle([x0, y0, x1, y1], outline="red", width=3)

    # class_id  
    draw.text((x0 + 2, y0 + 2), str(int(class_id)), fill="yellow")
```
